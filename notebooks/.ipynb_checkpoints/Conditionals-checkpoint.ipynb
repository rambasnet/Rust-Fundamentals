# Conditional Statements

- Conditional statements in Rust allow you to control the flow of your program based on certain conditions. The most common conditional statements are `if`, `else if`, and `else`.
- comparison operators are used to compare values and return a boolean result (`true` or `false`). These operators are essential for making decisions in your code.

## Comparison operators

| Operator | Description |
|----------|-------------| 
| `==`      | Equal to    |
| `!=`      | Not equal to |
| `<`       | Less than   |
| `>`       | Greater than |
| `<=`      | Less than or equal to |
| `>=`      | Greater than or equal to |

## One-way selection

- if expression
- The `if` expression evaluates a condition and executes a block of code if the condition is true. If the condition is false, the block of code is skipped.
- `( )` parentheses not required around the condition, but they can be used for clarity.

```rust
if condition {
    // code to execute if condition is true
}
```

````rust
let mut number = 99;
if number > 20 {
    println!("{number} is too big!");
}
````

````rust
if (number <= 20) {
    println!("{number} is less than or equal to 20");
}
````

## Two-way selection
- if-else expression
- The `if-else` expression allows you to specify an alternative block of code to execute if the condition is false.
```rust
if condition {
    // code to execute if condition is true
} else {
    // code to execute if condition is false
}
```

````rust
if number % 2 == 0 {
    println!("{number} is even.");
} else {
    println!("{number} is odd.");
}
````

````rust
number = 100;
if number % 2 == 0 {
    println!("{number} is even.");
} else {
    println!("{number} is odd.");
}
````

## Multiple-way selection
- if-else if-else expression
- The `if-else if-else` expression allows you to specify multiple conditions and corresponding blocks of code to execute based on which condition is true.
```rust
if condition1 {
    // code to execute if condition1 is true
} else if condition2 {
    // code to execute if condition2 is true
...
} else {
    // code to execute if both condition1 and condition2 are false
}
```

````rust
let number = 6;

if number % 4 == 0 {
    println!("number is divisible by 4");
} else if number % 3 == 0 {
    println!("number is divisible by 3");
} else if number % 2 == 0 {
    println!("number is divisible by 2");
} else {
    println!("number is not divisible by 4, 3, or 2");
}
````
